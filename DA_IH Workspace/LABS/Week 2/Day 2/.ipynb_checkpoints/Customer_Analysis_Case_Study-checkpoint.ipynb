```python
df3 = pd.concat([df1,df2],axis = 0, ignore_index = True)
```

```python
df3.columns = [i.lower().replace(" ","_") for i in df3.columns]
df3.head()
```

```python
list(df3.columns)
```

```python
df4 = df3.drop(columns = "customer")
```

```python
df4.head()
```

```python
#rearranging columns

df5 = df4[['st',
 'gender',
 'education',
 'customer_lifetime_value',
 'number_of_open_complaints',
 'policy_type',
 'vehicle_class',
 'income',
 'monthly_premium_auto',
 'total_claim_amount']]

df5.head()
```

```python
df5["number_of_open_complaints"].value_counts()
```

```python
df5['customer_lifetime_value'].value_counts()
```

```python
def func1(x):
    x = str(x)
    
    if x != "nan":
        y = x[:-1]
        return float(y)
    else:
        return 0

df5["customer_lifetime_value"] = df5["customer_lifetime_value"].apply(func1)
df5["customer_lifetime_value"].value_counts()
df5.head()
```

```python
def func2(x):
    x = str(x)
    
    if x.lower() == "nan":
        return 0
    else: 
        return int(x[2])
df5["number_of_open_complaints"] = df5["number_of_open_complaints"].apply(func2)
df5.head()
```

```python
type(df5['number_of_open_complaints'][0])
```

```python
#filtering data
df5['st'].unique()
```

```python
def state_cl(x):
    dct = {'Washington':'WA',
              'Arizona':'AZ',
               'Nevada':'NV',
               'California':'CA',
               'Cali':'CA',
               'Oregon':'OR',
               'AZ':'AZ',
               'WA':'WA',
                'NV':"NV"}
    if not x==x:
        #then I know that x is NaN
        return x
    else:
        return dct[x]
state_cl('NV')
```

```python
df5['st'] = df5['st'].apply(state_cl)
```

```python
#standardise the gender column
def gender_cl(x):
    dct = {'female':'F',
          'Male':'M',
          'Femal':'F',
           'F':'F',
           'M':'M'}
    if not x==x:
        #then I know that x is NaN
        return 'N'
    else:
    
        return dct[x]
gender_cl('Male')
```

```python
df5['gender'] = df5['gender'].apply(gender_cl)
```

```python
df5['region'] = [" "]*df5.shape[0]
df5.head()
```

```python
def region(x):  
    dct = {'CA':'West Region',
                 'WA':'East',
                 'OR':'North West',
                 'AZ':'Central',
                 'NV':'Central'}
    if x !=x:
        return x
    else:        
        return dct[x]
region('CA')
```

```python
df5['region'] = df5['st'].apply(region)
df5.head()
```
