```python
#Filtering data and Correcting typos â€“ Filter the data in state and gender column to standardize the texts in those columns
print(df2.gender.value_counts())
print(df2.state.value_counts())
```

```python
def state_cl(x):
    dct = {'California':'CA',
          'Oregon':'OR',
          'Arizona':'AZ',
           'Nevada':'NV',
           'Washington':'WA'}
    if not x==x:
        #then I know that x is NaN
        return x
    else:
        return dct[x]
state_cl('Nevada')
```

```python
df2.state = df2.state.apply(state_cl)
```

```python
#removing duplicates
df3 = df2.drop_duplicates()
df3.head()
```
