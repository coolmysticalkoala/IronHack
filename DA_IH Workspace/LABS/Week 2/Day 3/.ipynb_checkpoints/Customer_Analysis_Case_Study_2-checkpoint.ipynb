```python
df1 = df.drop(columns = "customer")
list(df1.columns)
```

```python
#rearranging columns

df2 = df1[['state',
           'gender',
           'location_code',           
           'education',
           'employment_status',
           'marital_status',           
           'response',
           'coverage',
          'policy_type',
           'policy',
           'renew_offer_type',
            'sales_channel',
           'vehicle_class',
           'vehicle_size',
           'vehicle_type',
            'number_of_policies',
            'number_of_open_complaints',
           'monthly_premium_auto',
           'months_since_last_claim',
           'months_since_policy_inception',
           'customer_lifetime_value',
           'income',
           'total_claim_amount',
           'effective_to_date']]

df2.head()
```

```python
#Filtering data and Correcting typos â€“ Filter the data in state and gender column to standardize the texts in those columns
print(df2.gender.value_counts())
print(df2.state.value_counts())
```

```python
def state_cl(x):
    dct = {'California':'CA',
          'Oregon':'OR',
          'Arizona':'AZ',
           'Nevada':'NV',
           'Washington':'WA'}
    if not x==x:
        #then I know that x is NaN
        return x
    else:
        return dct[x]
state_cl('Nevada')
```

```python
df2.state = df2.state.apply(state_cl)
```

```python
df2.head()
```

```python
#removing duplicates
df3 = df2.drop_duplicates()
df3.head()
```

```python
#Replacing null values â€“ Replace missing values with means of the column (for numerical columns)

df3._get_numeric_data()
```

```python
#turn to lower case
df4 = df3.applymap(lambda x: x.lower() if type(x) == str else x)
df4.head()
```

```python
df4.state = df3.state
df4.gender = df3.gender
```

```python
df4.head()
```

```python
#Replacing null values â€“ Replace missing values with means of the column (for numerical columns)
df4['number_of_policies'].fillna(value=df4['number_of_policies'].mean(), inplace=True)
df4['number_of_policies'].isnull().values.any()
```

```python
df4['monthly_premium_auto'].fillna(value=df4['monthly_premium_auto'].mean(),inplace=True)
df4['monthly_premium_auto'].isnull().values.any()
```

```python
'''Bucketing the data - Write a function to replace column "State" to different zones. 
California as West Region, Oregon as North West, and Washington as East, and Arizona and Nevada as Central'''

def region(x):  
    dct = {'CA':'West Region',
           'OR':'North West',
                 'WA':'East',                 
                 'AZ':'Central',
                 'NV':'Central'}
    if x != x:
        return x
    else:        
        return dct[x]
region('CA')
```

```python
df4['state'] = df4['state'].apply(region)
df4.head()
```

```python
'''Datetime format - Extract the months from the dataset and store in a separate column. 
Then filter the data to show only the information for the first quarter ,
ie. January, February and March. Hint: If data from March does not exist, consider only January and February.'''

from datetime import datetime
```

```python
df4['month'] = pd.to_datetime(df4['effective_to_date']).dt.strftime("%b")
df4.month.value_counts()
```
