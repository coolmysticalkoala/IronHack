## Webscraping exercise II

* Display the top 10 languages by number of native speakers stored in a pandas dataframe: url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'

```python
#1. importing libraries

import requests # to download html code
from bs4 import BeautifulSoup # to navigate through the html code
import pandas as pd
import numpy as np
import re
from time import sleep
from random import randint
```

```python
# 2. get url and store it in a variable

url = "https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers"
```

```python
# 3. download html with a get request. Use the function request.get() and store the output in response

response = requests.get(url)
# 200 status code means OK! response.status_code
print(response.status_code)
```

```python
# 4.1. parse html (create the 'soup')

soup = BeautifulSoup(response.text, 'html.parser')

# 4.2. check that the html code looks like it should

print(soup.prettify())
```

```python
#contents of the entire table under "wikitable sortable": #mw-content-text > div.mw-parser-output > table:nth-child(15)
```

```python
table = soup.find('table',{'class':'wikitable sortable'})
print(table)
```

```python
#return languages

res1 = soup.select("td:nth-child(2) > a")
print(res1)
```

```python
languages = []
for x in range(95):
    languages.append(res1[x].get('title'))
print(languages)
```

```python
# delete repeated languages (synonyms in the same column e.g. German and Standard German)

languages.remove("Standard German")
```

```python
languages.remove("Hindustani language")
```

```python
for x in range(95):
    num_speakers = [float(x.get_text()) for x in table.select('td:nth-of-type(3)')]
print(num_speakers)
```
